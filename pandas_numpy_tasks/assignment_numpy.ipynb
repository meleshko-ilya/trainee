**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
arr = np.random.randint(0, 10, size=10)
print(arr)
arr[(arr > 3) & (arr < 8)] *= -1
print(arr)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
arr3 = np.array([[1,2],[3,4]])
print("Matrix:n\n", arr3)
prod = np.outer(arr3.flatten(), arr3.flatten())
print("Outer product:\n", prod)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(0, 10, size=(8,3))
B = np.random.randint(0, 10, size=(2,2))
print("A matrix:\n", A)
print("\nB matrix:\n", B)
print("\nMatches:")
for i, b_row in enumerate(B):
    mask = np.array([all([elem in a_row for elem in b_row]) for a_row in A])
    if any(mask):
        print(A[mask])
                    
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
C = np.random.randint(0, 5, size=(10, 3))
print("C matrix:\n", C)
print("\nMethod 1:")
mask = np.apply_along_axis(lambda r: np.unique(r).size > 1, 1, C)
print(C[mask])
print("\nMethod 2:")
print(C[np.ptp(C, axis=1) != 0])
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
print("X matrix:\n", X)
# 1
res = 1
for i in range(min(len(X), len(X[0]))):
    if X[i][i] != 0:
        res *= X[i][i]
print(res)

# 2
print(np.diag(X))
print(np.prod(np.diag(X)[np.diag(X) != 0]))
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])
#1
values, counts = [], []
prev = x[0]
counter = 1
for i in x[1:]:
    if i == prev:
        counter += 1
    else:
        values.append(prev)
        counts.append(counter)
        prev = i
        counter = 1
values.append(prev)
counts.append(counter)
print("#1\n", values, counts)

#2
diff = np.diff(x)
idx = np.where(diff != 0)[0] + 1
blocks = np.split(x, idx)
values = np.array([v[0] for v in blocks])
counts = np.array([len(v) for v in blocks])
print("#2\n", values, counts)    
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import math
from scipy.spatial.distance import cdist
import time

X = np.random.randint(0, 10, size=(2,2))
Y = np.random.randint(0, 10, size=(2,2))
print("X:\n", X)
print("Y:\n", Y)
# 1
start_time = time.time()
result = []
for x in X:
    row = []
    for y in Y:
        d = math.sqrt(sum((xi - yi) ** 2 for xi, yi in zip(x, y)))
        row.append(d)
    result.append(row)
total_time = time.time() - start_time
print("#1\n", result)
print(total_time)
# 2
start_time = time.time()
result = np.sqrt(((X[:, None, :] - Y[None, :, :]) ** 2).sum(axis=2))
total_time = time.time() - start_time
print("#2\n", result)
print(total_time)
# 3
start_time = time.time()
result = cdist(X, Y)
total_time = time.time() - start_time
print("#3\n", result)
print(total_time)
```
